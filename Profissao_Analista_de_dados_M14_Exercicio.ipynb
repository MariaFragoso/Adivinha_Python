## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
tips = data[["cut", "price"]].groupby("cut").agg("sum").reset_index()
tips.head()
```

```
diamonds = data[["cut","price"]].groupby("cut").agg("mean").reset_index()
diamonds.head()
diamonds_graf = sns.barplot(data=diamonds, x="cut", y="price", ci=None, palette="pastel")
diamonds_graf.set(title='Preço por tipo diamante',ylabel='Preço',xlabel='Tipo');
```

---

Insight do gráfico 1: Podemos verificar a variação de preços,onde o diamante Premium é o mais caro e Ideal é o mais barato.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
flights = data[["year", "passengers"]].groupby("year").agg("sum").reset_index()
flights.head()
```

```
with sns.axes_style('whitegrid'):

  grafico = sns.lineplot(data=flights, x="year", y="passengers", palette="pastel")
  grafico.set(title='Passageiros por ano em Dezembro', xlabel='Ano', ylabel='Passageiros');
```

---

Insight do gráfico 2: Podemos verificar que o numero de passageiros foi aumento no mês de Dezembro.Além disso, o fato de que o número de passageiros aumenta em dezembro de cada ano pode sugerir que dezembro é um mês popular para viagens aéreas, possivelmente devido às férias e festividades de final de ano.


## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3
flights = data.query("1949 <= year < 1960")

with sns.axes_style('whitegrid'):

  grafico = sns.lineplot(data=flights, x="month", y="passengers", hue="year", palette="pastel")
  grafico.set(title='Passageiros por mês por ano', xlabel='Mês', ylabel='Passageiros');
  grafico.get_legend().set_title("Ano");
```

**Insight do gráfico 3: Percebe-se o aumento de passageiros ao longo dos anos, e o meses de Julho a Agosto registro um aumento ainda maior que o restante dos meses do ano.
